# [dbo].[v_batch_material_usage]

```sql
/****** Object:  View [dbo].[v_batch_material_usage]    Script Date: 06.05.2026 10:34:39 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- Расход сырья по партиям
CREATE VIEW [dbo].[v_batch_material_usage] AS
SELECT 
    pb.id AS batch_id,
    pb.batch_number,
    rm.name AS material_name,
    brm.planned_qty,
    brm.actual_qty,
    brm.unit,
    rmb.batch_number AS rm_batch_number,
    rmb.remaining_qty AS rm_remaining,
    s.name AS supplier_name
FROM production_batches pb
INNER JOIN batch_raw_materials brm ON pb.id = brm.batch_id
INNER JOIN raw_material_batches rmb ON brm.rm_batch_id = rmb.id
INNER JOIN raw_materials rm ON rmb.material_id = rm.id
LEFT JOIN suppliers s ON rmb.supplier_id = s.id;

GO
```

# [dbo].[v_step_executions_with_deviations]

```sql
/****** Object:  View [dbo].[v_step_executions_with_deviations]    Script Date: 06.05.2026 10:34:39 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- Выполнение техпроцесса с отклонениями
CREATE VIEW [dbo].[v_step_executions_with_deviations] AS
SELECT 
    se.id,
    pb.batch_number,
    ts.step_number,
    ts.name AS step_name,
    se.status,
    se.started_at,
    se.finished_at,
    se.actual_param_value,
    ts.param_name,
    ts.param_target,
    ts.param_min,
    ts.param_max,
    d.id AS deviation_id,
    d.severity,
    d.description AS deviation_description,
    DATEDIFF(MINUTE, se.started_at, se.finished_at) AS actual_duration_min
FROM step_executions se
INNER JOIN production_batches pb ON se.batch_id = pb.id
INNER JOIN tech_steps ts ON se.tech_step_id = ts.id
LEFT JOIN deviations d ON d.step_execution_id = se.id AND d.resolution_status = 'open';

GO
```

# [dbo].[v_production_order_summary]

```sql
/****** Object:  View [dbo].[v_production_order_summary]    Script Date: 06.05.2026 10:34:39 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- Сводка по производственным заказам
CREATE VIEW [dbo].[v_production_order_summary] AS
SELECT 
    po.id,
    po.order_number,
    p.name AS product_name,
    po.planned_qty,
    po.qty_unit,
    po.planned_date,
    po.status AS order_status,
    COUNT(DISTINCT pb.id) AS batch_count,
    SUM(pb.actual_qty) AS total_produced,
    AVG(DATEDIFF(HOUR, pb.started_at, pb.finished_at)) AS avg_production_hours
FROM production_orders po
LEFT JOIN products p ON po.product_id = p.id
LEFT JOIN production_batches pb ON po.id = pb.order_id
GROUP BY po.id, po.order_number, p.name, po.planned_qty, po.qty_unit, po.planned_date, po.status;

GO
```

# [dbo].[v_lab_tests_summary]

```sql
/****** Object:  View [dbo].[v_lab_tests_summary]    Script Date: 06.05.2026 10:34:39 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- Лабораторные испытания с результатами
CREATE VIEW [dbo].[v_lab_tests_summary] AS
SELECT 
    lt.id,
    lt.object_type,
    CASE 
        WHEN lt.object_type = 'raw_material_batch' THEN rmb.batch_number
        WHEN lt.object_type = 'production_batch' THEN pb.batch_number
    END AS batch_number,
    lt.status,
    lt.decision,
    lt.completed_at,
    u.full_name AS lab_user,
    COUNT(ltr.id) AS results_count,
    SUM(CASE WHEN ltr.is_within_range = 0 THEN 1 ELSE 0 END) AS out_of_spec_count
FROM lab_tests lt
LEFT JOIN raw_material_batches rmb ON lt.object_type = 'raw_material_batch' AND lt.object_id = rmb.id
LEFT JOIN production_batches pb ON lt.object_type = 'production_batch' AND lt.object_id = pb.id
LEFT JOIN users u ON lt.lab_user_id = u.id
LEFT JOIN lab_test_results ltr ON lt.id = ltr.test_id
GROUP BY lt.id, lt.object_type, rmb.batch_number, pb.batch_number, lt.status, lt.decision, lt.completed_at, u.full_name;

GO
```

# [dbo].[sp_complete_batch]

```sql
/****** Object:  StoredProcedure [dbo].[sp_complete_batch]    Script Date: 06.05.2026 10:34:39 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- Процедура для завершения партии
CREATE PROCEDURE [dbo].[sp_complete_batch]
    @batch_id INT,
    @actual_qty DECIMAL(12,3),
    @finished_by INT
AS
BEGIN
    SET NOCOUNT ON;
    BEGIN TRANSACTION;
    
    BEGIN TRY
        UPDATE production_batches
        SET 
            status = 'completed',
            actual_qty = @actual_qty,
            finished_at = GETDATE()
        WHERE id = @batch_id;
        
        -- Обновляем статус заказа, если все партии завершены
        UPDATE production_orders
        SET status = 'completed'
        WHERE id = (SELECT order_id FROM production_batches WHERE id = @batch_id)
          AND NOT EXISTS (
              SELECT 1 FROM production_batches 
              WHERE order_id = production_orders.id 
                AND status NOT IN ('completed', 'cancelled')
          );
        
        COMMIT TRANSACTION;
    END TRY
    BEGIN CATCH
        ROLLBACK TRANSACTION;
        THROW;
    END CATCH
END;

GO
```

# [dbo].[sp_consume_raw_material]

```sql
/****** Object:  StoredProcedure [dbo].[sp_consume_raw_material]    Script Date: 06.05.2026 10:34:39 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- =====================================================
-- 5. ХРАНИМЫЕ ПРОЦЕДУРЫ
-- =====================================================

-- Процедура для списания сырья
CREATE PROCEDURE [dbo].[sp_consume_raw_material]
    @batch_id INT,
    @rm_batch_id INT,
    @actual_qty DECIMAL(12,4),
    @added_by INT
AS
BEGIN
    SET NOCOUNT ON;
    BEGIN TRANSACTION;
    
    BEGIN TRY
        -- Добавляем запись о расходе
        INSERT INTO batch_raw_materials (batch_id, rm_batch_id, planned_qty, actual_qty, unit, added_by)
        SELECT 
            @batch_id, 
            @rm_batch_id, 
            planned_qty,
            @actual_qty,
            unit,
            @added_by
        FROM batch_raw_materials
        WHERE batch_id = @batch_id AND rm_batch_id = @rm_batch_id;
        
        -- Обновляем остаток сырья
        UPDATE raw_material_batches
        SET remaining_qty = remaining_qty - @actual_qty
        WHERE id = @rm_batch_id;
        
        -- Если остаток стал 0, меняем статус
        UPDATE raw_material_batches
        SET status = 'consumed'
        WHERE id = @rm_batch_id AND remaining_qty <= 0;
        
        COMMIT TRANSACTION;
    END TRY
    BEGIN CATCH
        ROLLBACK TRANSACTION;
        THROW;
    END CATCH
END;

GO
```
